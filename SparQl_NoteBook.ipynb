```sparql
%endpoint http://query.wikidata.org/sparql
%display table
%show all

PREFIX wikibase: <http://wikiba.se/ontology#>
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX wdt: <http://www.wikidata.org/prop/direct/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?president ?cause ?dob ?dod WHERE {
    ?pid wdt:P39 wd:Q11696 .
    ?pid wdt:P509 ?cid .
    ?pid wdt:P569 ?dob .
    ?pid wdt:P570 ?dod .

    OPTIONAL {
        ?pid rdfs:label ?president filter (lang(?president) = "en") .
    }
    OPTIONAL {
        ?cid rdfs:label ?cause filter (lang(?cause) = "en") .
    }
}
```

```sparql
%format json
%display table withtypes

# We might have more than one triple pointing to the same object, so we group by object
# and take one arbitrary predicate

  SELECT ?human ?humanLabel ?nativeName ?nativeNameLabel ?Oname ?OnameLabel ?deadPlace ?deadPlaceLabel ?dod ?dodLabel ?dob ?dobLabel ?birth_dat ?death_dat ?age WHERE {
  ?human wdt:P31 wd:Q5.
  ?human wdt:P1448 ?Oname.
  ?human wdt:P1559 ?nativeName.
#  FILTER(STR(?Oname) = 'Tommy Tabermann').
# ?human wdt:P570 ?dod.
# ?human wdt:P569 ?dob.    
  ?human p:P569/psv:P569 ?dob. 
  ?human p:P570/psv:P570 ?dod.
  ?dob wikibase:timeValue ?birth_date.
  ?dod wikibase:timeValue ?death_date.
        BIND( YEAR(?death_date) - YEAR(?birth_date) - 
              IF(MONTH(?death_date)<MONTH(?birth_date) || 
                 (MONTH(?death_date)=MONTH(?birth_date) && DAY(?death_date)<DAY(?birth_date)),1,0) AS ?age )
        # calculate the age, precisely to the day (times and timezones ignored)
        FILTER(?age > 18 && ?age < 100).    #ignore outlyers, several of which are probably errors
        FILTER(YEAR(?birth_date) > 1948).
        #FILTER(YEAR(?death_date) > 2017).
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
  
}
LIMIT 3
```
